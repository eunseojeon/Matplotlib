## ÌååÏù¥ Ï∞®Ìä∏ ÎßåÎì§Í∏∞
- PyplotÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Îã§Ïùå pie()Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏõêÌòï Ï∞®Ìä∏Î•º Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

- Î≥¥Îã§ÏãúÌîº ÏõêÌòï Ï∞®Ìä∏Îäî Î∞∞Ïó¥Ïùò Í∞Å Í∞í(Ïù¥ Í≤ΩÏö∞ [35, 25, 25, 15])Ïóê ÎåÄÌï¥ ÌïòÎÇòÏùò Ï°∞Í∞Å(ÏêêÍ∏∞ÎùºÍ≥† Ìï®)ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
- Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Ï≤´ Î≤àÏß∏ ÏêêÍ∏∞Ïùò ÌîåÎ°úÌåÖÏùÄ xÏ∂ïÏóêÏÑú ÏãúÏûëÌïòÏó¨ ÏãúÍ≥Ñ Î∞òÎåÄ Î∞©Ìñ•ÏúºÎ°ú Ïù¥ÎèôÌï©ÎãàÎã§ .

##ÎùºÎ≤®
- **labels** Îß§Í∞úÎ≥ÄÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏõêÌòï Ï∞®Ìä∏Ïóê Î†àÏù¥Î∏îÏùÑ Ï∂îÍ∞ÄÌï©ÎãàÎã§.
- **labels** Îß§Í∞úÎ≥ÄÏàòÎäî Í∞Å ÏêêÍ∏∞ÎßàÎã§ ÌïòÎÇòÏùò Î†àÏù¥Î∏îÏù¥ ÏûàÎäî Î∞∞Ïó¥Ïù¥Ïñ¥Ïïº Ìï©ÎãàÎã§.



```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

##ÏãúÏûë Í∞ÅÎèÑ
- ÏïûÏÑú Ïñ∏Í∏âÌñàÎìØÏù¥ Í∏∞Î≥∏ ÏãúÏûë Í∞ÅÎèÑÎäî xÏ∂ïÏóê ÏûàÏßÄÎßå **startangle** Í∞úÎ≥ÄÏàòÎ•º ÏßÄÏ†ïÌïòÏó¨ ÏãúÏûë Í∞ÅÎèÑÎ•º Î≥ÄÍ≤ΩÌï† Ïàò ÏûàÏäµÎãàÎã§.
- **startangle** Îß§Í∞úÎ≥ÄÏàòÎäî Í∞ÅÎèÑÎ°ú Ï†ïÏùòÎêòÎ©∞ Í∏∞Î≥∏ Í∞ÅÎèÑÎäî 0ÏûÖÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90) #Ï≤´Î≤àÏß∏ ÏêêÍ∏∞Î•º 90ÎèÑÏóêÏÑú ÏãúÏûë
plt.show()
```

## Explode
- Ïñ¥Îñ§ Ï°∞Í∞Å(wedge) ÌïòÎÇòÎ•º ÎèãÎ≥¥Ïù¥Í≤å ÌïòÍ≥† Ïã∂ÎÇòÏöî? explode ÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÎ©¥ Í∑∏Î†áÍ≤å Ìï† Ïàò ÏûàÏäµÎãàÎã§.
- explode ÌååÎùºÎØ∏ÌÑ∞Î•º ÏßÄÏ†ïÌïòÎ©¥, NoneÏù¥ ÏïÑÎãàÍ≥†, Í∞Å Ï°∞Í∞ÅÎßàÎã§ ÌïòÎÇòÏî© Í∞íÏùÑ Í∞ÄÏßÑ Î∞∞Ïó¥Ïù¥Ïñ¥Ïïº Ìï©ÎãàÎã§.
- Í∞Å Í∞íÏùÄ Í∞Å Ï°∞Í∞ÅÏù¥ Ï§ëÏã¨ÏóêÏÑú ÏñºÎßàÎÇò Îñ®Ïñ¥Ï†∏ ÌëúÏãúÎê†ÏßÄÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§.


```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode) #ÌååÏù¥ Ï§ëÏïôÏóêÏÑú "ÏÇ¨Í≥º" Ï°∞Í∞ÅÏùÑ 0.2ÎßåÌÅº ÎñºÏñ¥ÎÉÖÎãàÎã§.
plt.show()
```

##Í∑∏Î¶ºÏûê ÎÑ£Í∏∞
shadowsÎß§Í∞úÎ≥ÄÏàòÎ•º Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏÑ§Ï†ïÌïòÏó¨ ÏõêÌòï Ï∞®Ìä∏Ïóê Í∑∏Î¶ºÏûêÎ•º Ï∂îÍ∞ÄÌï©ÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"] #Í∞Å ÏêêÍ∏∞ÎßàÎã§ ÏÉàÎ°úÏö¥ ÏÉâÏÉÅÏùÑ ÏßÄÏ†ï

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:") #Ï†úÎ™©Ïù¥ ÏûàÎäî Î≤îÎ°Ä Ï∂îÍ∞Ä
plt.show()
```
