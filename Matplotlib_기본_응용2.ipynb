# ✅ Matplotlib 레이블 및 제목

## 플롯에 대한 레이블 만들기
- Pyplot을 사용하면 **xlabel()**및 **ylabel()**함수를 사용하여 x축과 y축에 대한 레이블을 설정할 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")  # Average Pulse → 평균 맥박
plt.ylabel("Calorie Burnage") # Calorie Burnage → 칼로리 소모량

plt.show()
```

### 플롯의 제목 만들기
- **title()**Pyplot을 사용하면 플롯의 제목을 설정하는 함수를 사용할 수 있습니다 .



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### 글꼴 지정 가능
- **xlabel(), ylabel(), title()**에서 **fontdict 파라미터**를 사용하면 제목과 레이블의 글꼴 속성을 설정할 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### 제목 위치 지정
- loc매개변수를 사용하여 title()제목의 위치를 ​​지정할 수 있습니다 .
- 유효한 값은 'left', 'right', 'center'입니다. 기본값은 'center'입니다.



```python
#제목을 왼쪽에 배치
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

### 플롯에 격자선 추가
- **grid()** Pyplot을 사용하면 플롯에 격자선을 추가하는 함수를 사용할 수 있습니다 .

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

### 표시할 격자선 지정
- axis함수의 grid() 매개변수를 사용하여 표시할 격자선을 지정할 수 있습니다.
- 유효한 값은 'x', 'y', 'both'입니다. 기본값은 'both'입니다.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

## 그리드의 선 속성 설정

## 여러개의 플롯 표시
- 이 **subplot()**기능을 사용하면 하나의 그림에 여러 개의 플롯을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

##산점도 만들기
- **scatter()**을 사용하면 산점도를 그릴 수 있습니다.
- 이 **scatter()**함수는 각 관측값에 대해 하나의 점을 표시합니다. 길이가 같은 두 개의 배열이 필요한데, 하나는 x축 값을 위한 것이고 다른 하나는 y축 값을 위한 것입니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

- 위의 예에서 관찰한 결과는 13대의 자동차가 지나간 결과입니다.
- X축은 자동차의 연식을 나타냅니다.
- Y축은 자동차가 지나갈 때의 속도를 보여줍니다.
- 자동차가 최신형일수록 더 빨리 달리는 것 같지만, 그것은 우연일 수도 있습니다. 결국 우리는 단 13대의 자동차만 등록했으니까요.

##플롯 비교
- 위의 예에서는 속도와 나이 사이에 관계가 있는 것처럼 보이지만, 다른 날의 관측치도 함께 표시해 보면 어떨까요? 산점도가 다른 정보를 알려주는 걸까요?

```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

- 아래 코드처럼 **color**또는 **c**인수를 사용하여 각 산점도에 대해 원하는 색상을 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

## ColorMap을 사용하는 방법
- 키워드 인수에 cmap컬러맵의 값을 지정하여 컬러맵을 지정할 수 있습니다. 이 경우 'viridis'Matplotlib에서 제공하는 기본 컬러맵 중 하나를 사용합니다.
- 또한 산점도의 각 지점에 대해 하나의 값, 즉 0~100까지의 값을 갖는 배열을 만들어야 합니다.

- 다음 문장 **plt.colorbar()**을 포함하면 도면에 색상 맵을 포함할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

## 점 크기 설정
- s인수를 사용하여 점의 크기를 변경할 수 있다.
- 색상과 마찬가지로 크기에 대한 배열의 길이가 x축과 y축에 대한 배열의 길이와 같은지 확인해보자.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## 점의 투명도 설정
- **alpha**인수를 사용하여 점의 투명도를 조정할 수 있습니다
- 색상과 마찬가지로 크기에 대한 배열의 길이가 x축과 y축에 대한 배열의 길이와 같은지 확인해보자.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## 색상, 크기, 알파 결합
- 다양한 크기의 점을 사용하여 컬러맵을 조합할 수 있음
- 점이 투명하면 가장 잘 보인다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
