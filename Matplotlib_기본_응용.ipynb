```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])  #Îã§Ïù¥Ïñ¥Í∑∏Îû®ÏóêÏÑú ÏúÑÏπò (0,0)ÏóêÏÑú ÏúÑÏπò (6,250)ÍπåÏßÄ ÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.

plt.plot(xpoints, ypoints)
plt.show()
```

## üìà Matplotlib ÌîåÎ°úÌåÖ

### x Î∞è y Ï†ê ÌëúÏãú
- Ïù¥ **plot()**Ìï®ÏàòÎäî Îã§Ïù¥Ïñ¥Í∑∏Îû®Ïóê Ï†ê(ÎßàÏª§)ÏùÑ Í∑∏Î¶¨Îäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.
- Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Ïù¥ **plot()**Ìï®ÏàòÎäî ÏßÄÏ†ê Í∞ÑÏóê ÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
- Ïù¥ Ìï®ÏàòÎäî Îã§Ïù¥Ïñ¥Í∑∏Îû®Ïùò ÏßÄÏ†êÏùÑ ÏßÄÏ†ïÌïòÍ∏∞ ÏúÑÌïú Îß§Í∞úÎ≥ÄÏàòÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.
- **Îß§Í∞úÎ≥ÄÏàò 1**ÏùÄ **xÏ∂ï**Ïùò Ï†êÏùÑ Ìè¨Ìï®ÌïòÎäî Î∞∞Ïó¥ÏûÖÎãàÎã§
- **Îß§Í∞úÎ≥ÄÏàò 2**Îäî **yÏ∂ï**Ïùò Ï†êÏùÑ Ìè¨Ìï®ÌïòÎäî Î∞∞Ïó¥ÏûÖÎãàÎã§
- (1, 3)ÏóêÏÑú (8, 10)ÍπåÏßÄ ÏÑ†ÏùÑ Í∑∏Î†§Ïïº ÌïòÎäî Í≤ΩÏö∞ plot Ìï®ÏàòÏóê Îëê Í∞úÏùò Î∞∞Ïó¥ [1, 8]Í≥º [3, 10]ÏùÑ Ï†ÑÎã¨Ìï¥Ïïº Ìï©ÎãàÎã§.

## üëèüèª ÌòïÏãù Î¨∏ÏûêÏó¥ **fmt**

- Îã®Ï∂ïÌÇ§ Î¨∏ÏûêÏó¥ ÌëúÍ∏∞Î≤ï Îß§Í∞úÎ≥ÄÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÎßàÏª§Î•º ÏßÄÏ†ïÌï† ÏàòÎèÑ ÏûàÏäµÎãàÎã§.
- Ïù¥ Îß§Í∞úÎ≥ÄÏàòÎäî ÎùºÍ≥†ÎèÑ Î∂àÎ¶¨Î©∞ fmtÎã§Ïùå Íµ¨Î¨∏ÏúºÎ°ú ÏûëÏÑ±Îê©ÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

- 'o' : Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º ÎèôÍ∑∏ÎùºÎØ∏(Circle) ÎßàÏª§Î°ú ÌëúÏãúÌï©ÎãàÎã§.
- ':' : Ï†êÏÑ†(dotted line) Ïä§ÌÉÄÏùºÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.
- 'r' : Îπ®Í∞ÑÏÉâ(red)ÏúºÎ°ú Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶ΩÎãàÎã§.
- Ï¶â, **'o:r'**Îäî **"Îπ®Í∞ÑÏÉâ Ï†êÏÑ† ÏúÑÏóê ÎèôÍ∑∏ÎùºÎØ∏ ÎßàÏª§Î°ú Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º ÌëúÏãú"**ÌïúÎã§Îäî ÎúªÏûÖÎãàÎã§.

## ÎßàÏª§ ÌÅ¨Í∏∞ ÏÑ§Ï†ï

- ÌÇ§ÏõåÎìú Ïù∏Ïàò **markersize**ÎÇò Îçî ÏßßÏùÄ Î≤ÑÏ†Ñ **ms**ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÎßàÏª§Ïùò ÌÅ¨Í∏∞Î•º ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20) #ÎßàÏª§ ÌÅ¨Í∏∞Î•º 20ÏúºÎ°ú ÏÑ§Ï†ï
plt.show()
```

## ÎßàÏª§ ÏÉâÏÉÅ

- ÌÇ§ÏõåÎìú Ïù∏Ïàò **markeredgecolor**ÎÇò Îçî ÏßßÏùÄ Ïù∏Ïàò **mec**Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÎßàÏª§ Í∞ÄÏû•ÏûêÎ¶¨ Ïùò ÏÉâÏÉÅÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

---
- ÌÇ§ÏõåÎìú Ïù∏Ïàò **markerfacecolor**ÎÇò Îçî ÏßßÏùÄ Ïù∏Ïàò **mfc**Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÎßàÏª§ Í∞ÄÏû•ÏûêÎ¶¨ ÎÇ¥Î∂ÄÏùò ÏÉâÏÉÅÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #ÎÇ¥Î∂Ä ÏÉâÏÉÅÏùÑ redÎ°ú ÏÑ§Ï†ï
plt.show()
```

---
- Ïù∏Ïàò **mecmfc** Î•º Î™®Îëê ÏÇ¨Ïö©ÌïòÏó¨ **ÎßàÏª§ Ï†ÑÏ≤¥**Î•º ÏÉâÏπ†Ìï©ÎãàÎã§.
