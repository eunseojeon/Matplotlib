```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])  #다이어그램에서 위치 (0,0)에서 위치 (6,250)까지 선을 그립니다.

plt.plot(xpoints, ypoints)
plt.show()
```

## 📈 Matplotlib 플로팅

### x 및 y 점 표시
- 이 **plot()**함수는 다이어그램에 점(마커)을 그리는 데 사용됩니다.
- 기본적으로 이 **plot()**함수는 지점 간에 선을 그립니다.
- 이 함수는 다이어그램의 지점을 지정하기 위한 매개변수를 사용합니다.
- **매개변수 1**은 **x축**의 점을 포함하는 배열입니다
- **매개변수 2**는 **y축**의 점을 포함하는 배열입니다
- (1, 3)에서 (8, 10)까지 선을 그려야 하는 경우 plot 함수에 두 개의 배열 [1, 8]과 [3, 10]을 전달해야 합니다.



```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10]) #다이어그램에서 위치 (1, 3)에서 위치 (8, 10)까지 선을 그립니다.

plt.plot(xpoints, ypoints)
plt.show()
```

- x축은 수평축
- y축은 수직축

## 🙅🏻‍♀️ 선 없이 플로팅

- 마커만 표시하려면 단축 문자열 표기 매개변수 'o'를 사용하면 되는데, 이는 '고리'를 의미합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10]) #다이어그램에 두 점을 그립니다. 하나는 위치 (1, 3)에, 다른 하나는 위치 (8, 10)에 있습니다.

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10]) #다이어그램에서 위치 (1, 3)에서 (2, 8)까지, 그 다음 (6, 1)까지, 마지막으로 위치 (8, 10)까지 선을 그립니다.

plt.plot(xpoints, ypoints)
plt.show()
```

## 👏🏻 형식 문자열 **fmt**

- 단축키 문자열 표기법 매개변수를 사용하여 마커를 지정할 수도 있습니다.
- 이 매개변수는 라고도 불리며 fmt다음 구문으로 작성됩니다.
- marker | line | color

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

- 'o' : 데이터 포인트를 동그라미(Circle) 마커로 표시합니다.
- ':' : 점선(dotted line) 스타일을 의미합니다.
- 'r' : 빨간색(red)으로 그래프를 그립니다.
- 즉, **'o:r'**는 **"빨간색 점선 위에 동그라미 마커로 데이터 포인트를 표시"**한다는 뜻입니다.

## 마커 크기 설정

- 키워드 인수 **markersize**나 더 짧은 버전 **ms**을 사용하여 마커의 크기를 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20) #마커 크기를 20으로 설정
plt.show()
```

## 마커 색상

- 키워드 인수 **markeredgecolor**나 더 짧은 인수 **mec**를 사용하여 마커 가장자리 의 색상을 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

---
- 키워드 인수 **markerfacecolor**나 더 짧은 인수 **mfc**를 사용하여 마커 가장자리 내부의 색상을 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #내부 색상을 red로 설정
plt.show()
```

---
- 인수 **mecmfc** 를 모두 사용하여 **마커 전체**를 색칠합니다.


```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

---
- 16진수 색상값을 사용할 수도 있음

## 선 스타일

- 키워드 인수 **linestyle**또는 shorter 버전인 **ls** 를 사용하여 그려진 선의 스타일을 변경할 수 있습니다.



```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed') #점선 활용
```

### 더 짧은 구문

- linestyle을 **ls**로 쓸 수 있습니다.
- dotted을 **:** 로 쓸 수 있습니다.
- dashed을 **--**로 쓸 수 있습니다.

```python
plt.plot(ypoints, ls = ':')
```

## 선 색상

- 키워드 인수 **color**나 더 짧은 인수**c**를 사용하여 선의 색상을 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r') #선 색상 빨간색
plt.show()
```

## 선 너비

- **lienwidth**키워드 인수나 shorter **lw**를 사용해 줄의 너비를 변경할 수 있습니다.
- 값은 포인트 단위의 부동 소수점 숫자입니다.



```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5') #20.5pt 폭의 선이 있는 플롯

plt.show()
```

## 여러줄 가능

- 더 많은 함수**plt.plot()**를 추가하면 원하는 만큼 많은 선을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```
